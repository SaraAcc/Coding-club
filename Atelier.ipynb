### 1.1. Les cellules
  
Jupyter est un environnement de programmation qui permet de faire de la programmation en Python. Il est composé de cellules qui peuvent être de deux types: des cellules de **texte** et des cellules de **code**.  
Les cellules de texte sont utilisées pour écrire du texte et les cellules de code sont utilisées pour écrire du code.    
Pour exécuter une cellule, il suffit de cliquer sur le bouton "Exécuter" dans la barre d'outils.

Tout d'abord, familiarisons nous avec la barre d'outils.  
  


Maintenant, voyons voir de quoi se compose une cellule !

- Code est le type d'une cellule de code
